There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. ## What NYC schools have the best math results?

1. Choosing only the needed columns

```python
best_math_schools = schools[['school_name', 'average_math']]
best_math_schools.head()
```

2. Calculating the math score in percent (total points is 800)

```python
best_math_schools['percent_score'] = best_math_schools['average_math'] / 800
best_math_schools.head()
```

3. Choosing the schools with scores of at least 80%

```python
best_math_schools = best_math_schools[best_math_schools['percent_score'] >= 0.8]
best_math_schools.head()
```

4. Selecting required columns and sorting the data frame

```python
best_math_schools = best_math_schools[['school_name','average_math']].sort_values('average_math', ascending=False)
best_math_schools.head()
```

## Which single borough has the largest standard deviation in the combined SAT score?

```python
schools_borough = schools
schools_borough.head()
```

1. Calculating the total SAT scores

```python
schools_borough['total_SAT'] = schools_borough['average_math'] + schools_borough['average_reading'] + schools_borough['average_writing']
schools_borough.head()
```

2. Calculating std of SAT scores for each borough
3. Sorting to see which borough has the largest std

```python
sat_std = schools_borough.groupby('borough')['total_SAT'].std()
sat_std = sat_std.sort_values(ascending=False)
sat_std
```

4. Extracting the value for Manhattan as it is the borough we are looking for

```python
manhattan_sat_std = sat_std[0]
manhattan_sat_std
```

5. Filtering the data frame

```python
manhattan_df = schools_borough[schools_borough['borough'] == 'Manhattan']
manhattan_df.head()
```

6. Calculating SAT scores mean

```python
manhattan_mean = manhattan_df['total_SAT'].mean()
manhattan_mean
```

7. Calculating the number od schools

```python
manhattan_num_schools = len(manhattan_df)
manhattan_num_schools
```

8. Combining the data into a single data frame
9. Setting the index to borough

```python
largest_std_dev = pd.DataFrame({'borough': ['Manhattan'],
                                'num_schools': [manhattan_num_schools],
                                'average_SAT': [manhattan_mean],
                                'std_SAT': [manhattan_sat_std]})

largest_std_dev.set_index('borough', inplace=True)
largest_std_dev.head()
```

10. Rounding the numbers

```python
largest_std_dev['average_SAT'] = largest_std_dev['average_SAT'].round(2)
largest_std_dev['std_SAT'] = largest_std_dev['std_SAT'].round(2)
largest_std_dev.head()
```
